### Question 1:  We'll always have Pareto

If X is a random variable with a Pareto distribution, then the probability that X is greater than some number x, i.e. the survival function (also called tail function), is given by:

$$
P(X > x) = \left(\frac{\alpha}{x}\right)^{\beta} \hspace{1cm}   x \ge \alpha\\
$$
$$
P(X > x)  = 1 \hspace{1cm}   x \lt \alpha
$$
 
where $\alpha$ is the (necessarily positive) minimum possible value of X, and $\beta$ is a positive parameter, which is known as the tail index.

Pareto originally used this distribution to describe how wealth is allocated amongst individuals, and the distribution of income. $\beta$ is used to describe this allocation ratio: the 80/20 rule where 20 percent of the people have 80 percent of the wealth corresponds to $\beta ~ 1.161$.

(a) Find analytically the Pareto CDF and show that the Pareto PDF is given by:

$$ P(x) =\frac{\beta \, \alpha^{\beta}}{x^{\beta+1}} \hspace{1cm} x\ge \alpha, \hspace{0.3cm} \alpha>0, \hspace{0.3cm} \beta>0  $$

Plot the PDF for $\beta=1.161,5$, and $\alpha=0.5,3$.

(b) Use the inverse method to to draw 10,000 samples from $X\sim {\rm Pareto}(\alpha=3, \beta=5)$. Plot a histogram of your samples.

(c) Suppose $X\sim {\rm Pareto}(\alpha=3, \beta=5)$. Use importance sampling to estimate $E[X]$ and 
$E[X^2]$ to within 3% error. Use as the importance function any pdf that you can find in `scipy.stats` (plot the pdf against the pareto to see which one is a good match). Generate 10,000 samples from the  pdf to obtain your estimates. 

Report your estimates for $E[X]$ and $E[X^2]$ along with the estimators errors, which are to be obtained by repeatedly calculating the monte-carlo integrals 100 times. Remember that these errors must be within 3%: this will require you to experiment with different pdfs.


**SOLUTION**

(a) The survival function is $s(x) = 1 - cdf(x)$ and $pdf(x) = \frac{d}{dx} cdf(x)$ and thus

$$pdf(x) = - \frac{d}{dx} s(x) = -\alpha^{\beta} (-\beta)x^{-(\beta+1)} = \frac{\beta \, \alpha^{\beta}}{x^{\beta+1}}$$

We plot the Pareto PDF for different values.

```python
alphas=[0.5, 3.]
betas=[1.161, 5.]

plt.figure()
for a in alphas:
    for b in betas:
        p = lambda x: b*(a**b)/(x**(b+1.))
        xvals=np.linspace(a, 10, 100)
        plt.plot(xvals, p(xvals), label="a %f b %f" % (a,b))
plt.legend()
plt.xlim([0, 10])
```

(b) Using the inverse method to draw from the Pareto:

We first plot the Pareto for $\alpha=3$ and $\beta=5$:

```python
alpha=3.
beta=5.
p = lambda x: beta*(alpha**beta)/(x**(beta+1.))
xvals=np.linspace(3, 10, 1000)
plt.plot(xvals, p(xvals), 'r')
```

To find the inverse we first note that the CDF of the Pareto can be obtained thus:

$$CDF(x) = \beta \alpha^{\beta} \int_{\alpha}^{x} \frac{du}{u^{\beta + 1}} $$

which is

$$ = 1 - \left(\frac{\alpha}{x}\right)^{\beta} $$

Thus if we draw from a uniform random variate on [0,1], $R$, corrsponding to the CDF, we have:

$$x = \alpha \, \left(1 - R\right)^{\frac{-1}{\beta}}$$ 

```python
alpha=3.
beta=5.
p = lambda x: beta*(alpha**beta)/(x**(beta+1.))

# CDF of p
CDF = lambda x: 1-(alpha/x)**beta

# invert the CDF
invCDF = lambda r: alpha*((1-r)**(-1./beta))

# domain limits
xmin = alpha # the lower limit of our domain
xmax = 100 # the upper limit of our domain

# range limits
rmin = CDF(xmin)
rmax = CDF(xmax)

N = 10000 # the total of samples we wish to generate

# generate uniform samples in our range then invert the CDF
# to get samples of our target distribution
R = np.random.uniform(rmin, rmax, N)
X = invCDF(R)

#hinfo = np.histogram(X,100)
# plot the histogram
plt.hist(X,bins=100, normed=True, label=u'Samples');

# plot our (normalized) function
xvals=np.linspace(xmin, xmax, 1000)
plt.plot(xvals, p(xvals), 'r', label=u'p(x)')
plt.xlim([0, 20])
# turn on the legend
plt.legend()
```

(c) We will estimate $E[X]$ and $E[X^2] $using importance sampling. Our proposal distribution, after a bit of experimentation
￼g(x) will be the gamma pdf: $g(x) = Gamma(a,b) = Gamma(15, 0.25)$ in the shape-scale form.
We will draw from this distribution for $x \ge \alpha$, and must thus renormalize the function over its support. The normalization is, obviously,

$$\frac{1}{1-CDFGamma(15,0.25)(\alpha)}$$. We can evaluate this at $x=\alpha=3$, the left side support we need for the pareto

```python
from scipy.stats import gamma
alpha=3.
beta=5.
p = lambda x: beta*(alpha**beta)/(x**(beta+1.))
q = lambda x: gamma.pdf(x,15, scale=0.25)
normize=1./(1. - gamma.cdf(3, 15, scale=0.25))
print normize
xvals=np.linspace(3, 10, 1000)
plt.plot(xvals, p(xvals), label="pareto")
plt.plot(xvals, normize*q(xvals), label="gamma")
plt.legend()
```

Solutions for $E[x]$ and $E[X^2]$ can ne calculated analytically for the pareto so we know the answer we want:

$$E[X] = \frac{\alpha\beta}{\beta-1}$$

and

$$E[X^2] = \frac{\alpha^2\beta}{\beta-2}$$

For $\alpha=3$ and $\beta=5$, this works out to 3.75 and 15 respectively.

Doing the importance sampling:


```python
N_samp=10000
N_sim=100
q = lambda x: normize*gamma.pdf(x,15, scale=0.25)
Isim=np.zeros(N_sim)
Isimx=np.zeros(N_sim)
Isimx2=np.zeros(N_sim)
for i in xrange(N_sim):
    xis = gamma.rvs(15, scale=0.25, size=N_samp)
    xis = xis[xis>=alpha]
    wratio = p(xis)/(normize*q(xis))
    Isim[i] =np.mean(wratio)
    Isimx[i] =np.mean(xis*wratio)
    Isimx2[i] =np.mean(xis*xis*wratio)
print "mean, std", np.mean(Isim), np.std(Isim)
print "E[x]", np.mean(Isimx)/np.mean(Isim)
print "error percent", 100*np.std(Isimx)/np.mean(Isimx)
print "E[x^2]", np.mean(Isimx2)/np.mean(Isim)
print "error percent", 100*np.std(Isimx2)/np.mean(Isimx2)
```

To find the CDF we integrate P(t’) from 0 to t:

Valentine's day is coming soon and you decided  to design a nice gift for your loved one.

Write a program that calculates the two moments of inertia of this sphere corresponding to rotation about the z and x axis. Do the calculation for a platinum  (21090 kg/m$^3$) sphere of radius 5 cm with an inner gold (19320 kg/m$^3$) cylinder of radius 1 cm.

where $r_{axis}(x, y, z)$ is the perpendicular distance of the point (x, y, z) from the axis of rotation. 
Calculate the values of $I_x$ and $I_z$.

**SOLUTION**

(a) Lets set it up:

```python
import time 
import math
densityOfPlatinum = 21090.00 #kg/m^3
densityOfGold = 19320.00 
platinumSphereRadius = .050 #m
goldCylinderRadius = .010 #m
boxMax = platinumSphereRadius 
boxMin = -boxMax
boxLength = boxMax - boxMin
```

We now calculate the moments of inertia for 1000 simulations and 10000 samples.

```python
boxVolume = boxLength**3 

numberOfSimulations = 1000
numberOfMonteCarloSamples = 10**4

#define a function of x, y, and z to give us the density
def densityFunctionOfWeirdObject(x, y, z):
    sphericalRadius = (x**2 + y**2 + z**2)**.5
    #if the point lies outside the sphere, automatically output 0 
    if sphericalRadius> platinumSphereRadius:
        return 0 
    else:
        #check to see if the point is within the z-aligned cylinder
        cylinderRadius = (x**2 + y**2)**.5
        if cylinderRadius > goldCylinderRadius:
            #if it is not, return the density of the sphere. otherwise of the cylinder
            return densityOfPlatinum 
        else:
            return densityOfGold
        
#define the integrand for the rotation about x
def IxIntegrand(x, y, z):
    distFromXSquared = z**2 + y**2
    return distFromXSquared * densityFunctionOfWeirdObject(x, y, z)
#define the integrand for the rotation about x
def IzIntegrand(x, y, z):
    distFromZSquared = x**2 + y**2
    return distFromZSquared * densityFunctionOfWeirdObject(x, y, z)

#vectorize to make faster
vectorizedIxIntegrand = np.vectorize(IxIntegrand) 
vectorizedIzIntegrand = np.vectorize(IzIntegrand)

startTime = time.time() 
calculatedIx =[] 
calculatedIz =[]

#doit evry simulation
for i in xrange(numberOfSimulations):
    randomNumbers = (boxMax - boxMin) * np.random.rand(3, numberOfMonteCarloSamples) + boxMin
    IxIntegrands = vectorizedIxIntegrand(randomNumbers[0], randomNumbers[1], randomNumbers[2])
    IzIntegrands = vectorizedIzIntegrand(randomNumbers[0], randomNumbers[1], randomNumbers[2])
    Ix = boxVolume * np.sum(IxIntegrands) / numberOfMonteCarloSamples
    Iz = boxVolume * np.sum(IzIntegrands) / numberOfMonteCarloSamples
    calculatedIx.append(Ix) 
    calculatedIz.append(Iz)
 
#calculate means
meanix=np.mean(calculatedIx) 
errorix=np.std(calculatedIx)
meaniz=np.mean(calculatedIz) 
erroriz=np.std(calculatedIz)
timeElapsed=time.time() - startTime

#print results
print "Ix = %s +/- %s kgm^2" % (meanix, errorix) 
print "Iz = %s +/- %s kgm^2" % (meaniz, erroriz) 
print "time elapsed is % seconds" % timeElapsed
```

(b) To pick a value of Nmc which gives an error of , we can use the formula for the generic error scaling of Monte Carlo to find

$$ 10^{-3} \sim \frac{1}{\sqrt(N_{mc})} \rightarrow N_{mc} \sim 10^{-6}$$

. One can get more accurate by calculating the ratio of the variance and the average as a function of monte carlo points and find the number of points for which this ratio is less of equal to 0.1%. In this way one finds $N_{mc} = 1.7 \times 10^{-6}$ 
which, as expected is of the same order of magnitude as the estimate from earlier. This number will vary some depending upon the accuracy of program implementation, etc. Note that this is the total number of samples, not just those accepted as being inside the sphere

```python
error=4 
Nx = 1657000
#first Ix
while error > .001:
    Nx=Nx+5000 #sufficient large step ensures quick decrease 
    randomNumbers = (boxMax - boxMin) * np.random.rand(3, Nx) + boxMin 
    IxIntegrands = vectorizedIxIntegrand(randomNumbers[0],
    randomNumbers[1], randomNumbers[2]) 
    avf = np.average(IxIntegrands)
    avfSquared = np.average(map(lambda x: x**2, IxIntegrands))
    error= ((avfSquared-avf**2)/Nx)**.5/avf 
    print Nx, error
print "Ix needed %s trials to get to error of less than .1%%" % str(Nx)
```

Generate a random sample from the below two dimensional

(b) In order to draw samples from this distribution, we create a box around the maximum and minimum values of and . Then, we draw random samples from this box and reject samples if . We find that for $N=2 \times 10^4$ samples we have $\frac{N+}{N-}$ about 30%.

```python
xmin=-2
xmax=2
ymin=-2
ymax=2
Pmin=0
Pmax=1
VBox=(xmax-xmin)*(ymax-ymin)*(Pmax-Pmin)
dN=1
N=2*10**(4)
Naccept=0
Nreject=0

niter=1000
Naccept=np.zeros(niter)
Nreject=np.zeros(niter)
for iter in range(niter):
    x1=(xmax-xmin)*np.random.rand(N)+xmin
    y1=(ymax-ymin)*np.random.rand(N)+ymin
    P1=(Pmax-Pmin)*np.random.rand(N)+Pmin
    Pt=np.exp(-(x1**2+np.cos(x1)*(y1**2)))
    Naccept[iter]=sum(P1<=Pt)
    Nreject[iter]=sum(P1>Pt)
print np.mean(Naccept)/np.mean(Nreject), np.std(Naccept)
```

(c) **Rejection Method on Steroids as an alternative**

We use

$$ \exp(\frac{-(x^2 + y^2)}{4}) $$

as the distribution function.  Now, we have reduced the volume to draw from and thus we reduce the number of samples that will be rejected

This function is simply a gaussian in 2D (radially in 1D actually but we can just use a multivariate gaussian)

```python
xmin=-2
xmax=2
ymin=-2
ymax=2
Pmin=0
Pmax=1
VBox=(xmax-xmin)*(ymax-ymin)*(Pmax-Pmin)
dN=1
N=100
N=2*10**(4)
Naccept=0
Nreject=0

niter=1000
Naccept=np.zeros(niter)
Nreject=np.zeros(niter)

#The ratio M, of the values at max, is 1.
for iter in range(niter):
    samps=np.random.multivariate_normal([0,0], [[4,0],[0,4]], N)
    xprop,yprop=samps.T
    xwhere = (xprop >= xmin)*(xprop <= xmax)
    ywhere = (yprop >= ymin)*(yprop <= ymax)
    mask=xwhere*ywhere
    newN=np.sum(mask)
    xpropnew=xprop[mask]
    ypropnew=yprop[mask]
    P1=(Pmax-Pmin)*np.random.rand(newN)+Pmin
    Pt=np.exp(-(xpropnew**2+np.cos(xpropnew)*(ypropnew**2)))
    Qt=np.exp(-(xpropnew**2+ypropnew**2)/4.0)
    ratio=Pt/Qt
    Naccept[iter]=sum(P1<=ratio)
    Nreject[iter]=sum(P1>ratio)
#print Naccept, Nreject
print np.mean(Naccept), np.mean(Nreject), np.std(Naccept)
```
