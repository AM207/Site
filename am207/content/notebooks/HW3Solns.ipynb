## Question 1

(a) Which part of the integrand is more suitable as the function to evaluate, and which is best used as a distribution to draw random variates from? Explain why.

We split the integrand as $f(x) = h(x)g(x)$. In choosing which part of the integrand is more
suitable to evaluate and which as a distribution from which to draw, we note that the distribution
should be easy to find a normalization factor for (not necessary in general, but in this case this is
possible and this way we don’t have to approximate the normalization say with Monte Carlo since
we can just get it analytically) and the new integrand should be ''flatter'' than the original one.
After some experimentation, we find an acceptable candidate is $g(x) = 1/[3+(x−3)^2]$. We can invert 
this analytically to sample the $x_i$ from the distribution g$*$, where $ g* = Cg $ for a normalization constant C.
Integrating from 0 to $ \infty $ we find that the normalization constant is $C = 6\sqrt{3}/5\pi$. Now we
￼can sample directly from g$*$ since integrating results in an easily invertible function (CDF). 

(b) Implement the Metropolis-Hastings algorithm that has as a stationary distribution to sample from, $g(x)$. Run the simulation 50 times for sample size of 150,000 and report the value of the integral I and the VAR(I).

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

g_norm = 6.0*np.sqrt(3.0)/(5*np.pi)
g = lambda x: g_norm/(3+(x-3)**2)
h = lambda x: x*np.log(x + 1)*np.exp(np.sqrt(x) - x)
f = lambda x: h(x)*g(x)
```

```python
x = np.arange(0, 14, 0.05)
plt.plot(x,g(x))
plt.plot(x,h(x))
plt.plot(x,f(x))
```

```python
gstar = lambda x: g_norm/(3+(x-3)**2) # target stationary distribution, appropriately normalized
propfun = lambda x: x + np.random.uniform(-1,1) # the proposal adds a realization drawn from U(-1,1) to the current point x
N = 1.5*10**5 # number of points
Ns = 50 # number of runs

IMH = np.zeros(Ns) # approximation for I, one for each run
```

```python
for i in np.arange(0,Ns):
    xvector = np.zeros(N) # for storing final settled on values at each step
    xprev = np.random.uniform(0,1) # random initial condition, may as well draw from U[0,1]. 
    unifvector = np.random.uniform(0,1,N) # vector to decide if we accept or reject proposed point

    for j in np.arange(0,N):
        xstar = propfun(xprev) #propose new value based on old one
        if xstar >= 0:
            if unifvector[j] <= min(1,gstar(xstar)/gstar(xprev)):
                xprev = xstar # move to new value
                xvector[j] = xprev #move to new value and store it
            else:
                xvector[j] = xprev # remain at old value and store it
        else:
            xvector[j] = xprev # remain at old value and store it
            
    xvector = xvector[0.2*N:] # burn first 20% of simulation
    IMH[i] = np.mean((1/g_norm)*xvector*np.log(xvector+1)*np.exp(np.sqrt(xvector)-xvector)) # approximation for I
    
print np.mean(IMH),np.var(IMH)
plt.hist(IMH)
```

(c) Now use accept-rejection method to draw from $g(x)$ and compare the two methods (compare I and VAR(I)). Which has the lowest variance? Why do you think this is the case?

```python
IAR = np.zeros(Ns) # approximation for I, one for each run
for i in np.arange(0,Ns):
    naccept = 0
    xaccept = np.zeros(N) # initialize acceptance vector
    while naccept < N:
        xtest = np.random.uniform(0,100) # uniform random number in [0,100] to approximate [0,infinity]
        ytest = np.random.uniform(0,0.3) # uniform random number in [0,.3], which is max of normalized g
        if ytest <= gstar(xtest): # if ytest is below curve
            xaccept[naccept] = xtest # value is accepted
            naccept += 1 # increment 
    IAR[i] = np.mean((1/g_norm)*xaccept*np.log(xaccept+1)*np.exp(np.sqrt(xaccept)-xaccept))

print np.mean(IAR),np.var(IAR)
plt.hist(IAR)
```

Using the rejection method to sample from $g*(x)$ we find that the variance is much smaller, or order $10^{−6}$. This method has a lower variance because we wait until we have all 150,000 points as a sample of $g*$, whereas for the Metropolis-Hastings algorithm we had to get rid of the first 20% of the samples for the burn-in.

We have an $m \times n$ matrix representing the pdf values in a 2-dimensional discrete space (saved as the file ‘pdf data.txt’). We implement a random walk Metropolis-Hastings algorithm to draw pairs from this distribution. We use component-wise updating. Since we draw from a uniform proposal, we use the Metropolis algorithm (and we manually restrict the ends so that we do not go off the grid, which introduces a slight error but does not prevent us from clearly seeing the message.

```python
import scipy as sp
import scipy.misc as spmisc

STEPS = 250000 # How many steps should we take
BURN_IN = 10000 # How many should we throw away for burn-in?

M,N = pdf.shape[0],pdf.shape[1] # How large is the array?

def mh_walk(step_sz):
    def proposal_dist(x,y):
        x_prop = x+sp.random.random_integers(-step_sz,step_sz)
        y_prop = y+sp.random.random_integers(-step_sz,step_sz)
        x_prop = sp.maximum(0,sp.minimum(x_prop,M-1))
        y_prop = sp.maximum(0,sp.minimum(y_prop,N-1))
        return (x_prop,y_prop)

    # Find starting value.
    x_values = sp.zeros(STEPS+1)
    y_values = sp.zeros(STEPS+1)

    x_values[0] = sp.random.random_integers(0,M)
    y_values[0] = sp.random.random_integers(0,N)

    for i in xrange(STEPS):
        x = x_values[i]
        y = y_values[i]
        x_prop,y_prop = proposal_dist(x,y)
        p_prop = pdf[x_prop][y_prop]
        p_curr  = pdf[x][y]

        a_p = (p_prop/p_curr) if (p_curr > 0) else sp.inf
        if sp.random.random() < a_p:
            x = x_prop
            y = y_prop
        x_values[i+1] = x
        y_values[i+1] = y

    # Remove burn-in.
    x_values = x_values[BURN_IN:]
    y_values = y_values[BURN_IN:]

    return (x_values,y_values)
```

```python
step_size=2   # Here we choose a step size = 2 to run our MH
mhout=mh_walk(step_size)
x_vals,y_vals = mhout[0],mhout[1]
imgout = sp.zeros((M,N))
for x,y in zip(x_vals,y_vals):
    imgout[x][y] += 1
imgplot = plt.imshow(imgout)
```

```python
step_size=25   # And now we choose a larger step = 25
mhout=mh_walk(step_size)
x_vals,y_vals = mhout[0],mhout[1]
imgout = sp.zeros((M,N))
for x,y in zip(x_vals,y_vals):
    imgout[x][y] += 1
imgplot = plt.imshow(imgout)
```

For part (d), we use the excellent answer by D. Newman as an example of how to invoke separate distributions for the x and y-direction steps.

```python
# Daniel begins by defining a multivariate MH function that 
#can handle two different functions for the proposal step.

def MH_multivariate_P3d(p,q,q2,n,gamma):
    # start from a random position 
    np.random.seed(seed=10)
    x0=0
    y0=0
    x_prev =x0
    y_prev =y0
    
    x = np.zeros(n)
    y = np.zeros(n)
    i=0
    accepted = 0
    
    while i<n:
        x_star =  q(x_prev) # specific x-direction proposal
        P_star = p(x_star,y_prev)
        P_prev = p(x_prev, y_prev)
        A =  P_star/P_prev 
        U = np.random.uniform();
        
        if (U < A):
            x[i] = x_star
            i = i + 1
            x_prev = x_star
            accepted+=1
        else: 
            x[i] = x_prev
            x_prev = x[i]
            i = i + 1
        
        i=i-1
        # now lets do the y step 
        y_star =   q2(y_prev) # specific y-direction proposal
        
        P_star = p(x_prev,y_star)
        P_prev = p(x_prev, y_prev)
        A =  P_star/P_prev
        
        U = np.random.uniform()
         
        if U < A:
            y[i] = y_star
            i = i + 1
            y_prev = y_star
            accepted+=1
        else: 
            y[i] = y_prev
            y_prev = y[i];  
            i = i + 1
            
    acceptance_rate = accepted/float(n*2)
    return acceptance_rate, x, y

# Here is the function we are going to sample from
def P(x1,x2):
    a = 0.1
    b = 1
    return np.exp((-1./(2*(a**2)))*((x1**2+x2**2)**0.5-1)**2-(1./(2*(b**2)))*((x2-1)**2))
```

Suppose we ask individuals to remember the chronological order of
  world cup winners. Individuals make a number of mistakes in the
  ordering of winners that can be captured by simple probabilistic
  models, such as Mallows model. Here's a simple
  example to explain Mallows model:
  
  Suppose that we are looking at the last four winners: France,
  Brazil, Italy, Spain.  We will represent this true ordering by a
  vector $\omega =$ (1,2,3,4) = (France, Brazil, Italy, Spain).
  Mallows model now proposes that the remembered orderings tend to be
  similar to the true ordering, with very similar orderings being more
  likely than dissimilar orderings. Specifically, according to Mallows
  model, the probability that an individual remembers an ordering
  $\theta$ is proportional to:

$ p(\theta | \omega, \lambda) \propto  e^{-\lambda d(\theta, \omega)} $

In this equation, $d(\theta, \omega)$ is the Kendall tau, $\tau$ is the
distance between two orderings. This distance measures the number of
adjacent pairwise swaps that are needed to bring the two orderings
into alignment.

For example, if $\theta = $ (Brazil, France, Italy,
Spain) then $d(\theta, \omega) = 1$ because one swap is needed to make
the two orderings identical. Similarly, if $\theta=$ (Brazil, Italy,
France, Spain), then $d(\theta, \omega) = 2$ because two swaps are
needed. Note that in Kendall tau distance, only adjacent items can be
swapped.  The scaling parameter $\lambda$ controls how sharply peaked
the distribution of remembered orderings is around the true
ordering. Therefore, by increasing $\lambda$, the model makes it more
likely that the correct ordering (or something similar) will be
produced.  The problem is now to generate orderings $\theta$ according
to Mallows model, given the true ordering $\omega$ and scaling
parameter $\lambda$. This can be achieved in very simple ways using a
Metropolis sampler.

Implement a Metropolis-Hastings algorithm to produce sample
guesses from 100 individuals. Choose various proposal distributions
and experiment with different values of $\lambda$.
