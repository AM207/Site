We define a function to plot the trace plot.

```python
# plot the trace plot
def plot_trace_plot(x_samples,y_samples,title):
    fig = plt.figure()
    plt.plot(np.arange(len(x_samples)),x_samples,label=r'$x$')
    plt.plot(np.arange(len(y_samples)),y_samples,label=r'$y$')
    plt.legend()
    plt.title(title + '- Trace Plot')
    plt.xlabel('Step')
    plt.ylabel('Trace')
    plt.xlim([0,5000]);
```

We define a function to calculate and plot autocorrelation,

```python
# define a function to calculate autocorrelation
def autocorr(k,arr):
    arr = arr - np.mean(arr)
    return (1./np.sum(arr**2))*np.sum((arr*np.roll(arr,-k))[:len(arr)-k])


# calculate and plot autocorrelation
def do_autocorr(x_samples,y_samples,title):
    # range of time lag in the autocorrelation calculation
    kplot = np.arange(501)

    # calculate autocorrelations
    autocorr_x = np.zeros(len(kplot))  
    for j1 in range(len(autocorr_x)):
        autocorr_x[j1] = autocorr(kplot[j1],x_samples)

    autocorr_y = np.zeros(len(kplot))
    for j2 in range(len(autocorr_y)):
        autocorr_y[j2] = autocorr(kplot[j2],y_samples)

    # plot autocorrelations
    fig = plt.figure()
    plt.plot(kplot[1:],autocorr_x[1:],label=r'$x$')
    plt.plot(kplot[1:],autocorr_y[1:],label=r'$y$')
    plt.legend()
    plt.title(title + '- Autocorrelation')
    plt.xlabel('Time Lag')
    plt.ylabel('Autocorrelation')
    plt.xlim([0,500]);
```

We define Geweke test, discarding the first 10 per cent of the samples.

```python
# define Geweke test
def Geweke(samples, intervals=20, length=200):

    # discard the first 10 per cent
    first = 0.1*len(samples)
    
    # create empty array to store the results
    z = np.empty(intervals)
    
    for k in np.arange(0, intervals):
        # starting points of the two different subsamples
        start1 = first + k*length
        start2 = len(samples)/2 + k*length
                
        # extract the sub samples
        subsamples1 = samples[start1:start1+length]
        subsamples2 = samples[start2:start2+length]
        
        # calculate the mean and the variance
        mean1 = np.mean(subsamples1)
        mean2 = np.mean(subsamples2)
        var1  = np.var(subsamples1)
        var2  = np.var(subsamples2)
        
        # calculate the Geweke test
        z[k] = (mean1-mean2)/np.sqrt(var1+var2)
    
    return z


# calculate and plot the Geweke test
def plot_Geweke(x_samples,y_samples,title):
    # perform the Geweke test
    Geweke_x = Geweke(x_samples)
    Geweke_y = Geweke(y_samples)
    
    # plot the results
    fig = plt.figure()
    plt.plot(Geweke_x,label=r'$x$')
    plt.plot(Geweke_y,label=r'$y$')
    plt.legend()
    plt.title(title + '- Geweke_Test')
    plt.xlabel('Subinterval')
    plt.ylabel('Geweke Test')
    plt.ylim([-3,3])
    
    # plot the delimiting line
    plt.plot( [2]*len(Geweke_x), 'r-.')
    plt.plot( [-2]*len(Geweke_y), 'r-.')
```

### Metropolis sampler

We define a componentwise Metropolis sampler,

```python
# define componentwise Metropolis algorithm
def metropolis(n_iterations, initial_values, prop_std):
    # initialize parameters
    trace = np.zeros((n_iterations+1,2))
    trace[0] = initial_values
    current_prob = func(*trace[0])

#------------------------------------------------------------------------------------
    for i in range(n_iterations):    
        # current parameters
        p = trace[i].copy()
        
        # perform componentwise Metropolis algorithm
        for j in range(2):
            # get current value for parameter j
            p[j] = rnorm(p[j],prop_std)
    
            # calculate pdf value at the proposed value
            proposed_prob = func(*p)

#-------------------------------------------------------------------------------------
            # sample a uniform random variate
            u = runif()
    
            # test proposed value
            if u < proposed_prob/current_prob:
                # accept
                trace[i+1,j] = p[j]
                current_prob = proposed_prob
            else:
                # stay put
                trace[i+1,j] = trace[i,j]
                p[j] = trace[i,j]

#---------------------------------------------------------------------------------------
    # return x-samples and y-samples separately
    x_samples = trace[1:,0]
    y_samples = trace[1:,1]

    # return results  
    return x_samples, y_samples
```

### Slice sampling

We define the slice sampler.

```python
# slice sampling
def slice_sampling(niter,pdf,x0,widths,dims=2):

    # initialize the sampler
    y0 = np.random.uniform(low=0,high=pdf(x0[0],x0[1]))
    samples = []

    # a hyperrectangle around the original point
    rectUnifs = np.random.uniform(size=dims)
    rectLefts = x0 - widths*rectUnifs
    rectRights = rectLefts + widths

#----------------------------------------------------------------------------------
    # do the sampling
    for i in range(niter):
        while (True):
            # new proposal
            xstarUnifs = np.random.uniform(size=dims)
            xstar = rectLefts + xstarUnifs*(rectRights - rectLefts)

            # loop until we find the point inside the region
            if y0 < pdf(xstar[0],xstar[1]):
                break
            else:
                # shrink rectangle to  make the search faster
                for j in range(dims):
                    if xstar[j] < x0[j]:
                        rectLefts[j] = xstar[j]
                    else:
                        rectRights[j] = xstar[j]
                
#--------------------------------------------------------------------------------------
        # save the current sample
        samples.append(xstar)

        # the last sample is now the proposal
        x0 = xstar

        # get the new y0 for next step
        y0 = np.random.uniform(low=0, high=pdf(x0[0],x0[1]))

        # reset the rectangle
        rectUnifs = np.random.uniform(size=dims)
        rectLefts = x0 - widths*rectUnifs
        rectRights = rectLefts + widths

#--------------------------------------------------------------------------------------
    # extract x-samples and y-samples
    x_samples = [s[0] for s in samples]
    y_samples = [s[1] for s in samples]
    
    x_samples = np.array(x_samples)
    y_samples = np.array(y_samples)
    
    return x_samples, y_samples
```

We run the chain and plot the two dimensional histogram.

```python
# perform slice sampling
x_samples, y_samples = slice_sampling(100000,func,[1,0],[1.0,1.0])

# plot the results
plot_2d_hist(x_samples,y_samples,'Slice Sampler')
```

We evaluate the trace plot. Compared to the Metropolis algorithm, the trace plot shows less autocorrelation. This is expected because we essentially accept all random walk in a slice sampler. However slice sampler is slower because (a) it is a data augmentation (i.e. simulating an additional dimension) and (b) we iterate the situation where the proposal is rejected.

```python
# plot the trace plot
plot_trace_plot(x_samples,y_samples,'Slice Sampler')
```

This is confirmed by the autocorrelations plot. Also note that, the autocorrelations are damped after about 100 steps, showing that the sampler is well mixed.

```python
# calculate and plot autocorrelations
do_autocorr(x_samples,y_samples,'Slice Sampler')
```

### Hamiltonian MC

First we define the Hamiltonian MC sampler.

```python
# define Hamiltonian MC
# consider the mass of the particle = 1
def Hamiltonian_MC(n_samples,x_init,epsilon,L):
    
    # the two dimensional Hamiltonian
    U = lambda x,y: -np.log(func(x,y))
    K = lambda px,py: (px**2+py**2)/2.

    # partial differentiation with respect to x and y
    dUdx = lambda x,y: x*(198.-200./np.sqrt(x**2+y**2))+2
    dUdy = lambda x,y: y*(200.-200./np.sqrt(x**2+y**2))+1

#------------------------------------------------------------------------------------
    # initialize the trajectory
    current_x = x_init[0]
    current_y = x_init[1]

    # the standard deviation follows the mass of the particle
    p_sig = 1.

    # initialize arrays to store results
    x_samples = np.zeros(n_samples)
    y_samples = np.zeros(n_samples)

#------------------------------------------------------------------------------------
    for j in xrange(n_samples):

        # the current position
        x = current_x
        y = current_y
    
        # draw a new momentum
        px = np.random.normal(0,p_sig)
        py = np.random.normal(0,p_sig) 
    
        # record this new momentum as the current momentum
        current_px = px
        current_py = py

#--------------------------------------------------------------------------------------
        # perform leap frog algorithm to calculate the trajectory
    
        # make a half step for momentum at the beginning
        px = px - epsilon*dUdx(x,y)/2.0
        py = py - epsilon*dUdy(x,y)/2.0
    
        # alternate full steps for position and momentum
        for i in xrange(L):
            x = x + epsilon*px
            y = y + epsilon*py
        
            if (i != L-1):
                px = px - epsilon*dUdx(x,y)
                py = py - epsilon*dUdy(x,y)

        # make a half step at the end
        px = px - epsilon*dUdx(x,y)/2.
        py = py - epsilon*dUdy(x,y)/2.
    
#------------------------------------------------------------------------------------------
        # decide whether we accept the current step
        current_U = U(current_x,current_y)
        current_K = K(current_px,current_py)
        proposed_U = U(x,y)
        proposed_K = K(px,py)
    
        A = np.exp(current_U - proposed_U + current_K - proposed_K)
    
        # accept/reject
        if np.random.rand() < A:
            current_x = x
            current_y = y
        
            x_samples[j] = x
            y_samples[j] = y
        else:
            x_samples[j] = current_x
            y_samples[j] = current_y
        
    return x_samples, y_samples
```

We run the Hamiltonian MC sampler and plot the two dimensional histogram.

```python
# run Hamiltonian sampling
x_samples, y_samples = Hamiltonian_MC(100000,[1,0],0.01,100)

# plot the results
plot_2d_hist(x_samples,y_samples,'Hamiltonian MC Sampler')
```

We evaluate the trace plot. The autocorrelation is even less than the slice sampler because we trace the dynamics closely in a Hamiltonian MC. However Hamiltonian MC sampler is the slowest because one need to integrate the trajectory.

```python
# plot the trace plot
plot_trace_plot(x_samples,y_samples,'Hamiltonian MC Sampler')
```

The autocorrelation is the lowest among all three methods.

```python
# calculate and plot autocorrelations
do_autocorr(x_samples,y_samples,'Hamiltonian Sampler')
```

We confirm the convergence using Geweke test.

```python
# calculate and plot the Geweke test
plot_Geweke(x_samples,y_samples,'Hamiltonian MC Sampler')
```

And finally, we perform a Gelman-Rubin test.

```python
print 'Hamiltonian Sampler- Gelman-Rubin Test'

# perform a Gelman-Rubin test
Gelman_Rubin_test('hamiltonian')
```

### Summary

### Cosmology
In the 1920s, Edwin Hubble, using the newly constructed 100" telescope at Mount Wilson Observatory, detected variable stars in several nebulae. Nebulae are diffuse objects whose nature was a topic of heated debate in the astronomical community: were they interstellar clouds in our own Milky Way galaxy, or whole galaxies outside our galaxy? This was a difficult question to answer because it is notoriously difficult to measure the distance to most astronomical bodies since there is no point of reference for comparison. Hubble's discovery was revolutionary because these variable stars had a characteristic pattern resembling a class of stars called Cepheid variables. Earlier, Henrietta Levitt, part of a group of female astronomers working at **Harvard College Observatory**, had shown there was a tight correlation between the period of a Cepheid variable star and its luminosity (intrinsic brightness). By knowing the luminosity of a source it is possible to measure the distance to that source by measuring how bright it appears to us: the dimmer it appears the farther away it is. Thus, by measuring the period of these stars (and hence their luminosity) and their apparent brightness, Hubble was able to show that these nebula were not clouds within our own Galaxy, but were external galaxies far beyond the edge of our own Galaxy.

Hubble's second revolutionary discovery was based on comparing his measurements of the Cepheid-based galaxy distance determinations with measurements of the relative velocities of these galaxies. He showed that more distant galaxies were moving away from us more rapidly:

$$v = H_0 d$$

where v is the speed at which a galaxy moves away from us, and d is its distance. The constant of proportionality Ho is now called the Hubble constant. The common unit of velocity used to measure the speed of a galaxy is km/sec, while the most common unitfor measuring the distance to nearby galaxies is called the Megaparsec (Mpc) which is equal to 3.26 million light years or 30,800,000,000,000,000,000 km! Thus the units of the Hubble constant are (km/sec)/Mpc.

The Hubble parameter measures the expansion rate at which the universe expands. Think about the Pythagoras theorem: it tells us that $ds^2 = dx^2 + dy^2$. This is true for a static space. But if space is expanding the the pythagoras theorem reads like $ds^2 = a(t) * (dx^2 + dy^2)$. $a(t)$ measures how fast the [angles of the] triangle is expanding with time t. This is what happens to the universe. The Hubble parameter simply measures the rate at which a(t) changes and is defined as:  $H(t)  = \frac{da(t)}{dt}\frac{1}{a(t)}$. Thus $H(t)$ tells us how fast space is growing. Astronomers measure it at different times to understand the nature of the universe.

The Hubble parameter can be obtained from the Friedmann-Robertson Walker equations as a function of the redshift $z$.  Cosmological redshift is seen due to the expansion of the universe, and sufficiently distant light sources (generally those more than a few million light years away) show redshift corresponding to the rate of increase in their distance from Earth.

```python
# Start with some definitions

H0_mean = 70.
H0_std = 10.
def log_prior_H0(H0):
    return scipy.stats.norm.logpdf(H0,loc=H0_mean,scale=H0_std)

Om_mean = 0.3
Om_std = 0.15
norm_prior_Om = scipy.stats.norm.cdf(1,loc=Om_mean,scale=Om_std) - scipy.stats.norm.cdf(0,loc=Om_mean,scale=Om_std)
def log_prior_Om(Om):
    return np.log( (Om>=0) * (Om<=1) * scipy.stats.norm.pdf(Om,loc=Om_mean,scale=Om_std) / norm_prior_Om )

def f(z,wk,H0,Om):
    return H0 * (z+1)**(3/2.) * np.sqrt( Om + (1-Om)*(z+1)**(3*wk) )

zs     = np.array([float(L.split()[0]) for L in data.split('\n')[1:-1]])
etas   = np.array([float(L.split()[1]) for L in data.split('\n')[1:-1]])
sigmas = np.array([float(L.split()[2]) for L in data.split('\n')[1:-1]])
def log_likelihood(wk,H0,Om):
    logp_sum = np.zeros(H0.shape)
    for (eta,z,sigma) in zip(etas,zs,sigmas):
        logp_sum += np.log(1./(sigma*np.sqrt(2*np.pi))) - ( eta - H0 * (z+1)**(3/2.) * np.sqrt( Om + (1-Om)*(z+1)**(3*wk) ) )**2 / (2*sigma**2)
    return logp_sum
def likelihood(wk,H0,Om):
    prod = np.zeros(H0.shape)
    for (eta,z,sigma) in zip(etas,zs,sigmas):
        prod += (1./(sigma*np.sqrt(2*np.pi))) * np.exp( - ( eta - H0 * (z+1)**(3/2.) * np.sqrt( Om + (1-Om)*(z+1)**(3*wk) ) )**2 / (2*sigma**2) )
    return prod

```

> Good enough. Notice that our prior for $\Omega_m$ clips slightly. This is intentional, since our domain is restricted to $[0,1]$. We've normalized it with its integral over that interval, so it's still a proper probability distribution (though we didn't need to, since we're just using MCMC).

> Now we construct the slice sampler itself.

```python
def PI(wk,sample): # target distribution
    return np.exp(log_likelihood(wk,*sample) + log_prior_H0(sample[0]) + log_prior_Om(sample[1]))

def H_slice_sampler(n=10000,burn=100,thin=1,initial=np.array([H0_mean,Om_mean]),window=np.array([30,0.3]), wk=-1):
    samples = np.zeros((n+burn,2))
    samples[0] = initial
    for i in xrange(1,n+burn):
        # Pick our augmenting point y (in logspace)
        aug_point = np.random.uniform(0,PI(wk,samples[i-1]))
        # Now create our hyperrectangle
        lower = np.maximum(np.array([0.,0.]), samples[i-1] - np.random.uniform(size=2)*window )
        upper = lower + window
        # Now sample and shrink
        candidate = np.random.uniform(size=(2))*(upper-lower) + lower
        while( aug_point>PI(wk,candidate) ):
            lower = np.maximum(np.array([0.,0.]), np.where( candidate<samples[i-1], candidate, lower ))
            upper = np.where( candidate>samples[i-1], candidate, upper )
            candidate = np.random.uniform(size=(2))*(upper-lower) + lower
        # Accept.
        samples[i] = candidate
    return samples[burn::thin]
```

2\. Calculate the Bayesian evidence ration for Model 1 and Model 2. Which model is more likely?

> The Bayesian evidence is defined as

$$ P(D|M_i) = \int p(D|\theta,M_i) p(\theta|M_i) \, d \theta $$

> In this case, our priors are independent of $M$, so we can use

$$ \int p(D|\theta,M_i)p(\theta)\,d\theta $$

> Which we can approximate using vanilla MC sampling:

$$ P(D|M_i) \approx \frac{1}{n}\sum_{\theta_k~p(\theta)} p(D|\theta_k,M_i) $$

> The Bayesian evidence ratio is defined as

$$ \frac{P(D|M_i)}{P(D|M_j)}
= \frac{\int p(D|\theta,M_i)p(\theta|M_i)\,d\theta}{\int p(D|\theta,M_j)p(\theta|M_j)\,d\theta}
= \frac{\int p(D|\theta,M_i)p(\theta)\,d\theta}{\int p(D|\theta,M_j)p(\theta)\,d\theta}
$$

> Substituting in our approximation, we get

$$
\frac{P(D|M_i)}{P(D|M_j)} \approx \frac{\sum_{\theta_k~p(\theta)} p(D|\theta_k,M_i)}{\sum_{\theta_k~p(\theta)} p(D|\theta_k,M_j)}
$$

> We can write code for that.

```python
def sample_H0_prior(n):
    return scipy.stats.norm.rvs(loc=H0_mean,scale=H0_std,size=n)
def sample_Om_prior(n):
    # rejection sample to deal with the clipped boundaries
    samples = np.zeros(n)
    n_sampled = 0
    while( n_sampled<n ):
        newsamps = scipy.stats.norm.rvs(loc=Om_mean,scale=Om_std,size=n-n_sampled)
        goodsamps = newsamps[np.logical_and(newsamps>0,newsamps<1)]
        n_good = goodsamps.shape[0]
        samples[n_sampled:n_sampled+n_good] = goodsamps
        n_sampled += n_good
    return samples
def bayes_factor(n=100000):
    # sample from priors
    H0 = sample_H0_prior(n)
    Om = sample_Om_prior(n)
    # compute likelihoods
    Like_M1 = likelihood(-1., H0, Om)
    Like_M2 = likelihood(-1.3, H0, Om)
    # return ratio
    return np.sum(Like_M1) / np.sum(Like_M2)
```

```python
%%time
bayes_factor()
```

> Model 1 is slightly more likely, but the Bayes' factor is pretty weak. Conventional wisdom says that the ratio should be at least an order of magnitude or more higher to be considered a strong argument.  We use as a reference the scale from Kass and Rafferty(1995) http://www.andrew.cmu.edu/user/kk3n/simplicity/KassRaftery1995.pdf:

* Bayes Factor 1 - 3.2  :  Not worth more than a bare mention
* Bayes Factor 3.2 - 10 :  Substantial
* Bayes Factor 10 - 100 :  Strong
* Bayes Factor > 100 :  Decisive
