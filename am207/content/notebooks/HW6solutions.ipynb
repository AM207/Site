### Question 1:  Squares beyond compare [10 pts]

Consider a shape consisting of two squares of equal side and centered at the same point, which 
 are rotated by $\pi/4$ relative to each other. The two squares intersect each other at eight points.
The squares each have four corners.  The intersections and corners lie on eight line 
segments (the sides of the two squares) that contain four points each, two corners and two intersections. 
Use Simulated Annealing to place the sixteen lowest positive integers along these 
line segments so that the sum of the numbers in each line segment is the same 
for all the line segments.
 



### Solution

Here, since we would like the sums of every four numbers to be the same, variance is a very natural choice for energy.


**Practical Choices**

- Starting temperature $T_0$: 10 (Note that I used $e^{-\frac{\Delta E}{T/5}}$ instead of $e^{-\frac{\Delta E}{T}}$ to adjust the acception rate. This is equal to adjust the temperature.)

- Lowering temperature schedule (thermostat): Linear - Temperature decreases as $c_{k+1}=αc_k$ with $α=0.99$.

- Reannealing interval: 100

- Stopping criterion: I checked all of the three criterions including Max iterations, Function tolerance (m=1000) and Objective limit (0.0, which means it has reached the optimal state).

- Transition jumps, L: $\sqrt{T}$

```python
def get_sums(numbers):
    qua_sums = []
    for i in xrange(0, 16, 2):
        idx = [i%16, (i+1)%16, (i+3)%16, (i+4)%16]
        qua_sums.append(sum(numbers[idx]))
    return qua_sums


def variance(numbers):
    # make list of qua-number sum
    qua_sums = get_sums(numbers)
    v = np.var(qua_sums)
    return v


def change_order(numbers, L):
    q = numbers.copy()
    for i in xrange(L):
        c1, c2 = np.random.choice(16, 2, replace=False)
        q[c1], q[c2] = q[c2], q[c1]
    return q
```

```python
def sa(numbers, init_temp, thermostat, ftol, itol, otol, reannealing):
    # ===============
    # SET THESE FOR DIAGNOSTICS
    # ================

    m = 1000
    vari = []
    temperature = init_temp

    it = 0                    

    # Set the variance as the energy
    prev_E = variance(numbers)
    atp=0
    while it >= 0:
        # NUMBER OF CORNERS IS L
        L = np.floor(np.sqrt(temperature)).astype(int)
        propose_numbers = change_order(numbers, L)

        new_E = variance(propose_numbers)
        deltaE  =  new_E - prev_E 

        if new_E < prev_E:
            numbers  = propose_numbers
            vari.append(new_E)
            prev_E = new_E  
            it = it + 1
        elif np.random.rand() < np.exp(-deltaE*5/temperature):
            numbers = propose_numbers
            vari.append(new_E)
            prev_E = new_E 
            it = it + 1

        atp =atp + 1;  # NUMBER OF ITERATIONS
        if it % reannealing == 0:
            temperature = thermostat * temperature;
            compl_temp = 0;
            if temperature < 1:
                temperature = 4

        if len(vari) > m and np.std(vari[-m:])/np.mean(vari[-m:]) < ftol:
            print 'ftol'
            break
        if atp > itol:
            print 'itol'
            break
        if len(vari) > 0 and vari[-1] <= otol:
            print 'otol'
            print vari[-1]
            break

    s = numbers
    return s, vari
```

### Solution

```python
# generate data
N = 100 # number of cities
cities = np.random.randint(1, 50, size=[2, N]) # generate the position of cities
pop = np.random.uniform(1, 10, N) # generate the population of cities
dist_thred = 10
pop_thred = 5
v_car = 1
v_air = 10

plt.plot(cities[0,:], cities[1,:], 'o')
plt.grid()
```

** a) Use Simulated Annealing to find solutions to the traveling salesman problem for 
$N=100$, optimizing the path by the total distance travelled while keeping track 
of the time of travel. **

First let's implement the helper functions and the simulated annealing algorithm. Here, learning from trials, I set the initial temperature as 20. In the algorithm, if the temperature is less than 2, I reset it to 8.

```python
def get_norm(a,b):
    sub = a - b
    return np.sqrt(np.sum(sub*sub))

def get_distance(pathcities):
    d=0.
    total = pathcities.shape[1]
    for j in xrange(total):
        d = d + get_norm(pathcities[:,j%total], pathcities[:, (j+1)%total])
    return d

def get_time(pathcities, pop, dist_thred, pop_thred, v_air, v_car):
    t = 0.
    total = pathcities.shape[1]
    for j in xrange(total):
        dist = get_norm(pathcities[:,j%total], pathcities[:, (j+1)%total])
        if dist >= dist_thred and pop[j] >= pop_thred and pop[(j+1)%total] >= pop_thred:
            t = t + dist/v_air
        else:
            t = t + dist/v_car
    return t
    

def changepath(inputcities, L):
    total = inputcities.shape[1]
    q = inputcities.copy()
    for i in xrange(L):
        c1, c2 = np.random.choice(total, 2, replace=False)
        temp = q[:, c1].copy()
        q[:, c1] = q[:, c2].copy()
        q[:, c2] = temp
    return q
```

```python
def sa(pathcities, pop, dist_thred, pop_thred, v_air, v_car, init_temp, thermostat, ftol, itol, otol, reannealing):
    # ===============
    # SET THESE FOR DIAGNOSTICS
    # ================

    m = 10000
    dist = []
    time = []
    temperature = init_temp
    
    it = 0                    

    # DISTANCE HERE IS OUR ENERGY 
    prev_E = get_distance(pathcities) 
    atp=0
    
    while it >=0:
        # NUMBER OF CORNERS IS L
        L = np.floor(np.sqrt(temperature)).astype(int)
        propose_path = changepath(pathcities, L)

        new_E = get_distance(propose_path)
        #print new_E
        deltaE  =  new_E - prev_E 

        if new_E < prev_E:
            pathcities  = propose_path
            dist.append(new_E)
            time.append(get_time(pathcities, pop, dist_thred, pop_thred, v_air, v_car))
            prev_E = new_E  
            it = it+1
        elif np.random.rand() < np.exp(-deltaE/temperature):
            pathcities  = propose_path
            dist.append(new_E)
            time.append(get_time(pathcities, pop, dist_thred, pop_thred, v_air, v_car))
            prev_E = new_E 
            it = it+1

        atp =atp +1;  # NUMBER OF ITERATIONS
        
        if it % reannealing == 0:
            temperature = thermostat * temperature;
            #temperature =  temperature/log(it);
            compl_temp = 0;
            if temperature < 2:
                temperature = 8
    
        if len(dist) > m and np.std(dist[-m:])/np.mean(dist[-m:]) < ftol:
            print 'ftol'
            break
        if atp > itol:
            print 'itol'
            break
        if len(dist) > 0 and dist[-1] <= otol:
            print 'otol'
            print dist[-1]
            break

    s = pathcities
    return s, dist, time
```

Now we run the algorithm.

```python
init_temp = 20
thermostat = 0.9
ftol = 0.01
itol = 20000
otol = 500
reannealing = 100

pcities, distances, times = sa(cities, pop, dist_thred, pop_thred, v_air, v_car, init_temp, thermostat, ftol, itol, otol, reannealing)

print "The path is ", pcities
```

Then we plot the traveling path along with the distances and times vs iteration. In the traveling path, the pathes by airplane is labeled as red.

```python
plt.figure()
plt.plot(pcities[0,:], pcities[1,:], 'o')
for i in xrange(100):
    dist = get_norm(pcities[:,i], pcities[:, (i+1)%100])
    if dist >= dist_thred and pop[i] >= pop_thred and pop[(i+1)%100] >= pop_thred:
        plt.plot([pcities[0,i], pcities[0, (i+1)%100]], [pcities[1,i], pcities[1, (i+1)%100]], color='r')
    else:
        plt.plot([pcities[0,i], pcities[0, (i+1)%100]], [pcities[1,i], pcities[1, (i+1)%100]], color='k')

plt.title("Traveling Path")
plt.grid()

plt.figure()
plt.plot(distances)
plt.grid()
plt.title("Distance")

plt.figure()
plt.plot(times)
plt.grid()
plt.title("Time")
```

As we can see, both distance and time dicrease along with iterations. Loosely, they follow the same pattern, i.e. they go up and down together. In this problem setting, according to several trials I conducted, the minimal distance can approach very close to 800. But since we have a large number of cities, the result varies every time running, and it is hard to find the global optimum, although theoretically SA will guarantee one if we run it long enough.

** b) Now redo the problem by optimizing the the path by the total time of travel, while keeping track of the distance traveled. Are the two solutions similar or different? **

Again, let's first modify the above simulated annealing algorithm to adapt it to problem (b).

```python
def sa2(pathcities, pop, dist_thred, pop_thred, v_air, v_car, init_temp, thermostat, ftol, itol, otol, reannealing):
    # ===============
    # SET THESE FOR DIAGNOSTICS
    # ================

    m = 10000
    dist = []
    time = []
    temperature = init_temp
    
    it = 0                    
    
    # time is energy 
    prev_E = get_time(pathcities, pop, dist_thred, pop_thred, v_air, v_car) 
    atp=0
    
    while it >=0:
        # NUMBER OF CORNERS IS L
        L = np.floor(np.sqrt(temperature)).astype(int)
        propose_path = changepath(pathcities, L)

        new_E = get_time(propose_path, pop, dist_thred, pop_thred, v_air, v_car)
        deltaE  =  new_E - prev_E 

        if new_E < prev_E:
            pathcities  = propose_path
            time.append(new_E)
            dist.append(get_distance(pathcities))
            prev_E = new_E  
            it = it+1
        elif np.random.rand() < np.exp(-deltaE/temperature):
            pathcities  = propose_path
            time.append(new_E)
            dist.append(get_distance(pathcities))
            prev_E = new_E 
            it = it+1

        atp =atp +1;  # NUMBER OF ITERATIONS
        
        if it % reannealing == 0:
            temperature = thermostat * temperature;
            #temperature =  temperature/log(it);
            compl_temp = 0;
            if temperature < 2:
                temperature = 8

        if len(dist) > m and np.std(dist[-m:])/np.mean(dist[-m:]) < ftol:
            print 'ftol'
            break
        if atp > itol:
            print 'itol'
            break
        if len(dist) > 0 and dist[-1] <= otol:
            print 'otol'
            print dist[-1]
            break

    s = pathcities
    return s, dist, time
```

Now we run the algorithm.

```python
init_temp = 20
thermostat = 0.9
ftol = 0.01
itol = 20000
otol = 500
reannealing = 100

pcities, distances, times = sa2(cities, pop, dist_thred, pop_thred, v_air, v_car, init_temp, thermostat, ftol, itol, otol, reannealing)
```

(2) In the travelling path, we can see clearly that by optimizing the total time, there are more travels by flight. This makes sense because faster speed gives shorter time.

The results show interesting patterns. First, with the distance as the objective function, we can see that actually different population thresholds don't influence the optimization result at all because the objective function (distance) doesn't depend on population threshold. This can also be seen and justified in the travling path figures in that there are equal amount of "red lines" (flight travels). But we can see an obvious decrease in total time since the higher the population threshold is, the fewer flights there are, thus the longer the traveling time will be.

On the other hand, if we use the total time as the objective function, there is an obvious difference with difference thresholds. With a very small population threshold, almost all of the travels can be taken by flight and the average traveling time (energy in SA) is small. With the current setting of temperature, almost all of the proposed results during the SA process will be accepted. Therefore it is very hard realize the optimization with the current paramter settings. Then as the threshold increases, the average traveling time increases and there are fewer chances that we will accept the proposed result. This is why we have less data points in the red line and the purple line. But we can see an obvious pattern of optimization in them. (Notice that the larger the threshold is, the more consistent distance and time will be. This is why the distance is also more optimized with larger threshold.) Also, the final value increases as the threshold increases because with fewer and fewer available flights, the traveling time will increase. Such results can also be seen in the traveling path figures. The smaller the threshold is, the more red lines there are, and the less compact/optimized the path is.
