### Question 1: The perfect configuration

As we were advised to do in lab, I used a hexagonal starting configuration.  I didn't worry about choosing good locations for type $A$ and $B$ particles, because I figured the algorithm would sort that out.  The starting configuration is plotted below.

```python
population = 50
N = 19
particles = np.zeros((population,2,N))
N_A = 7

linDist = 1.1
offsetDist = np.sqrt(linDist**2 - (linDist/2)**2)

x = [-2*linDist, -linDist, 0, linDist, 2*linDist]
x2 = [-1.5*linDist, -0.5*linDist, 0.5*linDist, 1.5*linDist]
particles[:,0,0:3] = x[1:4]
particles[:,1,0:3] = offsetDist*2
particles[:,0,3:7] = x2
particles[:,1,3:7] = offsetDist
particles[:,0,7:12] = x
particles[:,1,7:12] = 0.0
particles[:,0,12:16] = x2
particles[:,1,12:16] = -offsetDist
particles[:,0,16:19] = x[1:4]
particles[:,1,16:19] = -offsetDist*2

plt.figure(figsize=[8,8])
plt.scatter(particles[0,0,:N_A], particles[0,1,:N_A], c='r', s=50, label='A')
plt.scatter(particles[0,0,N_A:], particles[0,1,N_A:], c='b', s=50, label='B')
plt.title('Starting configuration')
plt.legend()
plt.show()
```

Below are the utility functions for the genetic algorithm.  The $\text{populateMatrices}$ function just creates matrices that are used to determine which $\sigma$ and $\epsilon$ to use for each pair of points.  The first 7 particles in the array are always type $A$; if type $A$ and $B$ particles need to be swapped, the coordinates are swapped, not the types.

The $\text{V}$ function calculates the Lennard-Jones potential given $\epsilon$, $\sigma$, and $r_{12}$ for a pair of points.

The $\text{dist}$ function calculates the Euclidean distance $(r_{12})$ between a pair of points.

The $\text{getV}$ function calculates the Lennard-Jones potential for the entire configuration for every individual in the population.  It does this by iterating through each pair of points in the configurations and summing all the Lennard-Jones potentials.

The $\text{mutate}$ function can move (jiggle) a particle slightly, and it can swap a type $A$ particle with a type $B$ particle.  Both of these mutations happen with relatively low probability.  I tried to optimize the mutation probabilities to yield the best result most often.

I also tried recombination by picking two random individuals, picking a random particle in one configuration, picking the closest particle in the second configuration, and swapping the coordinates.  The idea is that different configurations will be similar, so by choosing the closest point in the second configuration, the two configurations will be changed, but not too much.

```python
sigmaA = 1.0
sigmaB = 1.125
sigmaAB = np.sqrt(sigmaA*sigmaB)
epsA = 2.0
epsB = 1.5
epsAB = (epsA+epsB) / 2

typeA = [i < N_A for i in xrange(N)]  # type of each particle, True = A

sigmaMatrix = np.zeros((N,N))
epsMatrix = np.zeros((N,N))

def populateMatrices():
    for i in range(N):
        for j in range(i+1,N):
            if typeA[i] and typeA[j]:
                sigmaMatrix[i,j] = sigmaMatrix[j,i] = sigmaA
                epsMatrix[i,j] = epsMatrix[j,i] = epsA
            elif not typeA[i] and not typeA[j]:
                sigmaMatrix[i,j] = sigmaMatrix[j,i] = sigmaB
                epsMatrix[i,j] = epsMatrix[j,i] = epsB
            else:
                sigmaMatrix[i,j] = sigmaMatrix[j,i] = sigmaAB
                epsMatrix[i,j] = epsMatrix[j,i] = epsAB

V = lambda eps,sigma,r12: eps * ( (sigma/r12)**12 - (sigma/r12)**6 )

dist = lambda x1,x2,y1,y2: np.sqrt( (x1-x2)**2 + (y1-y2)**2 )

def getV(particles):
    vsum = np.zeros(population)
    for i in range(N):
        for j in range(i+1,N):
            r12 = dist(particles[:,0,i], particles[:,0,j], particles[:,1,i], particles[:,1,j])
            vsum += V(epsMatrix[i,j], sigmaMatrix[i,j], r12)
    return vsum

def mutate(particles):
    jiggleMutate = 0.03
    swapTypeMutate = 0.03

    # jiggle mutations
    allRand = np.random.rand(population,2,N)
    idx = allRand < jiggleMutate
    add = np.random.normal(0, 0.01, np.sum(idx))
    particles[idx] += add

    # swap A/B mutations
    allRand = np.random.rand(population)
    idx = allRand < swapTypeMutate
    numSwaps = np.sum(idx)
    idxA = np.random.choice(N_A, numSwaps, replace=False)
    idxB = np.random.choice(xrange(N_A,N), numSwaps, replace=False)
    particles[idx,:,idxA], particles[idx,:,idxB] = particles[idx,:,idxB], particles[idx,:,idxA]

    # select two individuals for recombination
    indv = np.random.choice(population, 2, replace=False)
    idx = np.random.randint(0, N)
    idx2 = np.argmin(dist(particles[indv[0],0,idx], particles[indv[1],0,:],
                          particles[indv[0],1,idx], particles[indv[1],1,:]))
    particles[indv[0],:,idx], particles[indv[1],:,idx2] = particles[indv[1],:,idx2], particles[indv[0],:,idx]
```

The following code cell performs the genetic optimization.  Each individual in the population goes through a mutation phase, which may or may not actually change anything about that individual.  Then the Lennard-Jones potential is computed for each configuration.  The LJ potential is converted into a fitness value by the function $e^{-V}$, where $V$ is the LJ potential for a configuration.  I used $e^{-V}$ for two reasons.  First, it differentiates between similar values, such as $-16$ and $-18$.  These values are not much different percentage-wise, but after computing $e^{-V}$, they become much different, and that's good from an evolutionary perspective, because I want $-18$ to survive at the expense of $-16$.  The other reason I used $e^{-V}$ is that it eliminates all negative values, which could gum up the cumulative sum function.

After computing the fitness for each configuration, I check to see if the maximum fitness is better than the current best, and if it is, I save that configuration.

Then I compute the cumulative sums of the fitness values and map those to the interval $[0,1]$.  I generate one random value for each individual and find the smallest cumulative sum that is greater than that value, and the associated individual gets to have a baby.  I create a new population from the old one, duplicating the individuals that had offsping.

```python
generations = 4000
maxFitness = -1e100
idxs = np.zeros(population, dtype=np.int32)
fitnessHist = np.zeros(generations)

populateMatrices()

for i in xrange(generations):
    mutate(particles)
    vs = getV(particles)
    fitnesses = np.exp(-vs)
    maxIdx = np.argmax(fitnesses)
    fitnessHist[i] = fitnesses[maxIdx]
    if fitnesses[maxIdx] > maxFitness:
        maxFitness = fitnesses[maxIdx]
        bestConfig = particles[maxIdx].copy()
    cumsum = np.cumsum(fitnesses) / np.sum(fitnesses)
    randoms = np.random.rand(population)
    for j in xrange(population):
        idxs[j] = np.argmax(cumsum > randoms[j])
    particlesNew = np.zeros((population,2,N))
    for j in xrange(population):
        particlesNew[j] = particles[idxs[j]]
    particles = particlesNew
```

The maximum fitness (the fitness of the most fit individual) at each generation is plotted below.  The optimized configuration is also shown.  The type $A$ particles should end up forming the innermost hexagon because the optimal $A$-$A$ distance is slightly smaller than the optimal $A$-$B$ and $B$-$B$ distances.  Ideally, the $B$ particles at the corners of the outer hexagon will be pushed out slightly, so that the hexagon looks like it has concave sides.  This is because the corner particles are pushed out by the two neighboring $B$ particles, while the $B$ particles on the sides are pulled in just a bit by the $A$ particles, because the optimal $A$-$B$ distance is less than the optimal $B$-$B$ distance.

Unfortunately the algorithm doesn't find the perfect configuration every time it runs, but it generally gets very close.  This particular run found a very nice looking configuration.

```python
plt.figure(figsize=[17,8])
plt.subplot(1,2,1)
plt.plot(fitnessHist)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Maximum fitness at each generation')
plt.subplot(1,2,2)
plt.scatter(bestConfig[0,:N_A], bestConfig[1,:N_A], c='r', s=50, label='A')
plt.scatter(bestConfig[0,N_A:], bestConfig[1,N_A:], c='b', s=50, label='B')
plt.title('Optimized configuration')
plt.legend()
plt.show()
```

### Question 2: The cost of housing

#### Question 2a: Least-squares cost function

We are trying to minimize $\|X\theta-y\|_2$ $($or equivalently $\|X\theta-y\|_2^2)$, where $X$ is the matrix with columns $x^{(i)}_0=1$, $x^{(i)}_1=\text{Area}^{(i)}$, and  $x^{(i)}_2=\text{Bedrooms}^{(i)}$.  The vector $y$ is the right hand side, where $y^{(i)} = \text{Price}^{(i)}$.

So the cost function $J(\theta)$ can be defined as:

$$ J(\theta) = \sqrt{\sum_{i=1}^m [f_{\theta}(x^{(i)}) - y^{(i)}]^2} $$

where $f_{\theta}(b) = \theta^T b$ for a vector $b$.  This cost function is equivalent to the 2-norm of the residual.

#### Question 2b: Batch gradient descent

The following code cell opens the data file (which must be in the working directory of the iPython notebook), and reads the data into a 2D array.  The data is then split into an $X$ matrix containing the areas and bedrooms and a $y$ vector containing the prices.  A column of $1$s is added to the $X$ matrix to allow an additive constant $\theta_0$.  The areas and bedrooms are normalized by dividing by the minimum value of each.

The $\text{getCost}$ function calculates the 2-norm of the residual of $\theta$, which is $\|X\theta-y\|_2$

The $\text{getGradient}$ function calculates $-\nabla J_i(\theta) = [y^{(i)} - f_{\theta}(x^{(i)})] \cdot 2x^{(i)}$

```python
table = np.loadtxt("houses.txt", delimiter=',', dtype=np.float64)

def normalize(b):
    norm = np.min(b)
    b /= norm
    return norm

X = table[:,0:2]
y = table[:,2]
numRecords = y.size
X = np.insert(X, 0, np.ones(numRecords), axis=1)  # add a column of 1s to matrix

getGradient = lambda x,y,theta: (y-np.dot(theta,x)) * 2*x
getCost = lambda X,y,theta: np.linalg.norm(np.dot(X,theta) - y)

thetaReal, realResidual, _, _ = numpy.linalg.lstsq(X, y)
realResidual = np.sqrt(realResidual)

areaNorm = normalize(X[:,1])
bedsNorm = normalize(X[:,2])
```

The $\text{batchDescent}$ function below implements the batch gradient descent algorithm.  The $\theta$ vector is initialized to values that are in the general neighborhood of the actual least-squares solution.  For a set number of iterations, the function computes the sum of the gradients at each data point, and it adds $\alpha \sum_{i=1}^m -\nabla J_i(\theta)$ to the existing $\theta$, where $-\nabla J_i(\theta)$ is returned by the $\text{getGradient}$ function.

I tried multiple values for $\alpha$, and many of them diverged, including $\alpha=1.3\times 10^{-3}$.  Clearly with $\alpha$ too large, the algorithm steps too far each time.  So it steps past the solution and ends up farther from the solution then where it started.  I plotted only $\alpha$ values that converged.  I ended up using $\alpha=1.1\times 10^{-3}$ because that value yielded the best convergence after 50 iterations, even though it was not the best after 5 iterations.

```python
def batchDescent(X,y,N,alpha):
    theta = np.array([100000.0, 100000.0, -7500.0])
    costs = []
    for i in xrange(N):
        sumGrad = np.zeros(3)
        for j in xrange(numRecords):
            sumGrad += getGradient(X[j], y[j], theta)
        theta += sumGrad * alpha
        costs.append(getCost(X,y,theta))
        if costs[-1] - realResidual < 0.1:
            break
    return theta,costs

plt.figure(figsize=[8,6])
alphas = [0.0012, 0.0011, 0.001, 0.0005, 0.0001, 0.00005]
for alpha in alphas:
    theta,costs = batchDescent(X,y,50,alpha)
    plt.plot(costs, label=r'$\alpha=%.1e$' % alpha)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title(r'Cost per iteration for different $\alpha$ values')
plt.legend()
plt.show()
```

The following code cell runs the batch descent algorithm to convergence and prints the results.  I defined convergence as getting a residual $\|X\theta-y\|_2$ within $0.1$ of the residual from the least squares computation.  I picked $0.1$ because both algorithms were able to achieve this in a reasonable amount of time, and they were both within $100 of the correct answer on part (e).

The optimized $\theta$ and cost are shown below, along with the elapsed time in seconds and a plot of the cost per iteration.

```python
import time
start = time.time()
thetaBatch,costsBatch = batchDescent(X,y,999999999,0.0011)
end = time.time()
thetaBatch[1] /= areaNorm
thetaBatch[2] /= bedsNorm

def printResults(theta, minCost, costs, time):
    print "Optimized theta_0: %.3f" % theta[0]
    print "Optimized theta_1: %.3f" % theta[1]
    print "Optimized theta_2: %.3f" % theta[2]
    print "Optimized cost: %.3f" % minCost
    print "Elapsed seconds for gradient descent: %.3f" % time
    plt.figure(figsize=[8,6])
    plt.semilogx(costs)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.title(r'Cost per iteration')

printResults(thetaBatch, costsBatch[-1], costsBatch, end-start)
plt.show()
```

#### Question 2c: Stochastic gradient descent

The $\text{stochasticDescent}$ function below implements the stochastic gradient descent algorithm.  I plotted the performance of different values of $\alpha$, but when running to convergence I used the same $\alpha$ value that I used in the batch descent, just to make sure the comparison was fair.

```python
def stochasticDescent(X,y,N,alpha):
    theta = np.array([100000.0, 100000.0, -7500.0])
    costs = []
    minCost = getCost(X,y,theta)
    minTheta = theta.copy()
    for i in xrange(N):
        order = np.random.permutation(numRecords)
        for j in order:
            grad = getGradient(X[j], y[j], theta)
            theta += grad * alpha
        costs.append(getCost(X,y,theta))
        if costs[-1] - realResidual < 0.1:
            break
    return theta,costs

plt.figure(figsize=[8,6])
alphas = [0.0012, 0.0011, 0.001, 0.0009, 0.0008, 0.0007]
for alpha in alphas:
    _, costs = stochasticDescent(X,y,50,alpha)
    plt.plot(costs, label=r'$\alpha=%.1e$' % alpha)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title(r'Cost per iteration for different $\alpha$ values')
plt.legend()
plt.show()
```

The following code runs the stochastic descent algorithm to convergence and prints the results.  Convergence is defined the same way as for the batch descent algorithm.

The optimized $\theta$ and cost are shown below, along with the elapsed time in seconds and a plot of the cost per iteration.

```python
start = time.time()
thetaStoch, costsStoch = stochasticDescent(X,y,999999999,0.0011)
end = time.time()
thetaStoch[1] /= areaNorm
thetaStoch[2] /= bedsNorm
printResults(thetaStoch, costsStoch[-1], costsStoch, end-start)
plt.show()
```

#### Question 2d: Algorithm comparison

The stochastic descent algorithm took longer to converge, which was somewhat surprising.  I can see a couple reasons for this.  Probably the most important is that the batch algorithm is always converging toward the best answer, while the stochastic algorithm is bouncing around.  That bouncing can cost extra time.  The other reason is that the stochastic algorithm multiplies by $\alpha$ at each data point, while the batch algorithm multiplies by $\alpha$ once per iteration.  I'm guessing this is not a huge cost.

The results of the two algorithms are fairly similar, which is not surprising since I required both algorithms to get within $0.1$ of the actual best 2-norm residual.

The stochastic algorithm did outperform the batch algorithm for a few iterations, as can be seen in the plot below.  The fact that the stochastic algorithm required more iterations to converge indicates that the bouncing around is probably the biggest contributor to its longer running time.

**Instructor's Note**: Most students found that the stochastic algorithm took longer to converge than the batch gradient descent algorithm. As noted in lecture, stochastic gradient descent is useful to get a quick and dirty approximation for a solution.  One reason that students found the stochastic algorithm converged more slowly, is that the convergence condition was rather stringent.  With a less stringent convergence condition, the stochastic algorithm might appear to reach a solution more quickly.  The runtime differences between the two algorithms will also likely be greater for more complicated cost functions and problems with large datasets.
