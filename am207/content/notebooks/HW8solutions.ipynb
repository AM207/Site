####(a) Using Yule Walker

For an AR(3) model,

$$x_t = A_{1}x_{t-1}+ A_{2}x_{t-2} +A_{3}x_{t-3} + z_t.$$

If we multiply both sides of this equation by $x_{t-1}$ $x_{t-2}$ and $x_{t-3}$ respectively, take ensemble expectations, and divide by the variance of x, we get k equations:

\begin{eqnarray}
\rho_1 &=& A_{1} + A_{2}\rho_1 + A_{3}\rho_{2}\\
\rho_2 &=& A_{1}\rho_1 + A_{2} + A_{3}\rho_{1}\\
\rho_3 &=& A_{1}\rho_{2} + A_{2}\rho_{1} + A_{3}
\end{eqnarray}

We get the Yule-Walker equations. Denote the matrix on the right hand side as $R$, we have

$$ \vec{A} = (A_{1}, A_{2}, A_{3})^T $$

$$ \mathbf{R} =  \left( \begin{array}{ccc}
1 & \rho_1 & \rho_2 \\
\rho_1 & 1 & \rho_1 \\
\rho_2 & \rho_1 & 1 \end{array} \right)  $$

$$ \vec{\rho} = (\rho_{1}, \rho_{2}, \rho_{3})^T $$

Solving equation $\mathbf{R}\vec{a} = \vec{\rho}$ by Cramer's rule, we will get the parameters $ \vec{A} $ as we want.

```python
rho = sm.tsa.stattools.acf(data, nlags=3)[1:] # calculate autocorrelation
rho
```

```python
R = [[1, rho[0], rho[1]], [rho[0], 1, rho[0]], [rho[1], rho[0], 1]]
A = np.linalg.solve(R, rho) # solve linear equations
print "Parameters estimated using Yule-Walker are: ", A
```

#### Comparing yule-walker to mle

So whats going on?

The 'nc' here only makes sure we dont add a constant. This is not really relevant here as we constructed the data set to be constant-less, but its important to know what you are using in your fits.

A critical thing to remember is that YW and MLE are not identical. The former matches moments, the latter does a conditional regression on the last p coeffs for an AR(p) series. In other words, the autocovariances in YW are estimated using all the data in the time series, but in MLE this is not the case. As the time series becomes longer these converge. So the results will be close, but will be identical only for infnite N, whereas we only have 100.

Since our time series is short and seems to be in a realization where persistence of structure is relatively high at the beginning, the effect of initial conditions doesn't become negligible. YW or Least Squares conditional on initial conditions will not be the same as full MLE that assumes the initial conditions follow the stationary distribution.

####(b) Roots of the backshift Polynomial


For AR(2) process

$$ X_t =  \alpha_1 X_{t-1} + \alpha_2 X_{t-2} + z_t $$

The corresponding function in terms of the backshift operator $B$ is

$$  (1-\alpha_1 B - \alpha_2 B^2) \, x_t = z_t $$

Lets look at the roots of 

$$ (1- \alpha_1 B - \alpha_2 B^2) = 0 $$

we can write the equation as 

$$ (1-g_1 B) (1-g_2 B) =0 $$ 

The roots are 

$$ 1/g_1, 1/g_2 = \frac{-\alpha_1 \pm \sqrt{\alpha_1^2+4\alpha_2}}{2\alpha_2}$$

With $\alpha_1 = 1.1664$, $\alpha_2 = -0.6399$,

$$ 1/g_1 = 0.9114-0.8556j $$
$$ 1/g_2 = 0.9114+0.8556j $$
$$ \vert g_1 \vert = 0.799937497558 < 1 $$
$$ \vert g_2 \vert = 0.799937497558 < 1 $$

Therefore, the process is stationary. It can also be shown through

\begin{eqnarray} 
   \alpha_1 + \alpha_2 = 0.5265 &<&1 \\
   -\alpha_1 + \alpha_2 = -1.8063 &<&1 \\
   -1< \alpha_2 = -0.6399 &<&1 \\
\end{eqnarray}

One way to think about why we are able to obtain periodic behavior from the AR(2) process because the roots for equation $ (1- \alpha_1 B - \alpha_2 B^2) =0 $ are complex. Another way to think about it is that the signs of the coefficients are opposite one another.

Also, one can use:

####(c) Finding the best order using the AIC

```python
orders = range(20)
aics = np.zeros(len(orders))
for o in orders:
    aro = sm.tsa.ARMA(lynx, (o,0)).fit(method="mle")
    aics[o] = aro.aic
```

```python
print "The best fit order is", np.argmin(aics)
print "The corresponding AIC is ", min(aics)

plt.plot(orders, aics, '-o')
plt.title("AIC vs AR Model Order")
plt.xlabel('order')
plt.ylabel('AIC')
```
