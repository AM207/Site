# Mr. Anderson and AM207: A preview of what's coming. i.e. How AM207 will help you survive more than interviews and job talks


The story begins in 1981. Mr Anderson was traveling on a twin engine plane for the richness of South America. His brother, Jeffrey, has made a lot of money finding gold in the Andean mountains. Jeffrey was the adventurous type; he dropped out of high school and followed a friend who was going down to South America. He returned 10 years later a very rich man. Mr Anderson on the other hand stayed at school and then went to one of the most prestigious universities in North America -- Harvard University -- where he studied computational science under the famous Prof Provlief. These studies never got Mr Anderson any money, fame or even popularity. He was a loner, shy and very uncomfortable with people he did not know. What nowadays people might call a nerd, back then people was termed "peculiar".

After graduating from Harvard, he took odd jobs at different places in Boston but none of them satisfied his curiosity or his intellect.


When he heard that the mining companies in the desert of Atacama were looking for bright young people to uncover the riches of nitrite (and having the requisite very good understanding of engineering and mathematics) he decided to take the risk and go down there to make his fortune.

After a long trip to Lima, Peru, he boarded a twin engine that would take him to San Pedro in the middle of the desert, the driest place on earth; even lizards dreamed of spa treatments out there. Half way through the trip they were crossing a high pass of the Andes above the desert. He could look down from his window upon the infinite desert.

It was then that he realized that one of the engines was on fire. He informed the pilot, who told him not to worry as they could continue on the other engine. But all of the sudden the other engine stopped operating. The pilot was quick to jump out of the plane with the only chute that was available. Ready to die, Anderson sat on the seat and quickly came to the realization that life was all about luck. He remembered the words of his professor telling him about **randomness**.
<br>

<br>
The plane went down very fast and luckily landed on the soft sand (lizards and spa treatments will do that to the sand). Mr. Anderson had survived despite the impossible odds. **"There is always that small probability and if you do not simulate things properly, you will miss those outliers"**, he remembered his wise professor telling him. "My professor was so right", he thought.

He got out of the plane, and dusted himself off.  Happily concluding that nothing was broken and that he had survived a plane crash, he started examining the surroundings. There was absolutely nothing around. No water, no food and it was still hot and dry. "I will not survive", he thought. The voice of his professor was in his head though. "There is always that small little probability, the world is **ergodic** and if you search hard enough and smart enough you can get everywhere in your search space." Right about now, he wished that he knew what ergodic meant. 

He took out his handy old GPS device (an old version of GPS) and started walking west hoping he could get to the coast. He knew the coast was hundred of kilometers away but he would try to use any information he had to his benefit. He also took with him all medical supplies and whatever else he could get his hands on that seemed useful.

After 3 hours of trekking exhausted, thirsty and about to collapse, he saw something that looked like a water. He ran and indeed, there was a little pond.

'How lucky I am', he thought. He was ready to drink when he realized that there was a sign "Poison, do not drink". He looked into the medical supplies for he remembered there was something resembling an antidote for poisonous water.

He got it out and read the instructions. One drop per cubic meter of water. Do not use more. Excess use can have severe side effects, such as blindness. Mr. Anderson has to calculate the volume of the water. Luckily the pond seem to have a constant depth of about the length of his palm, which he knew was 30 cm (because even though he grew up in the US using imperial units, when in South America, do as the South Americans do and use metric!). He knew he had to somehow calculate the volume of the water in the pond. How though? Now he remembered a trick that he learned back in Harvard under Professor Provlief. First he formed a square on the sand using his body as a measuring stick. Next he(as randomly as he could) threw stones onto the square. By measuring the number of stones that landed in the water vs the total number of stones he had thrown, he was able to calculate the area of the pond and therefore the volume of the water.

$A_{pond} = \frac{N_{in}}{N_{total}} A $




He continued his westward direction, however all of sudden a sand storm started and visibility became very limited; he could not see anything now and had hard time breathing.

<div id="picture"><img width="400" src="files/images/desert-survival-8.jpg">
</div>
    

He knew if he could go to the highest point on the landscape he would have a much better chance in surviving. But without seeing the task of finding the highest point was impossible. Luckily again, his old professor had taught him how to find the maximum in a parameter space, which seemed to be perfectly appropriate for his situation. Remembering his professor's explanation, he took a few steps in a random direction and determined if his vertical position went up or down. If he stepped downwards he reversed his position and tried again, otherwise he continued from the new position. He remembered and incorporated from his old lectures a point about being willing every now and then to accept bad positions (i.e. going downwards vertically) to avoid being stuck at local maxima. Following this procedure, three hours later he was able to find the high ground and saved his life once more.

He stayed at high ground for a while until the sand storm passed. Unfortunately, now it was cloudy and he could not see the sun. During the sand storm he lost his sense of directionality. He still had the old GPS but also found a pedometer that could provide him with estimates of his speed and with some basic astronomy he knew he could approximately find the direction. Starting to walk westward he faced the dilemma: should he trust the old GPS or should he trust the pedometer and his estimations? All of sudden he remembered Prof. Provlief lectures on Kalman filters. He remembered that this was a way to find out which one to believe more. Using the Kalman's gain he was able to maintain course and after a few more hours of walking he saw the far lights of a little village.


$ \hat{X}_k = K_k \, Z_k +(1-K_k) \hat{X}_{k-1} $

"I am saved!!" he thought. He started running towards the lights but when he arrived closer, he realized that this was not civilization as he was hoping for. This was a small dwelling out in the country with a few people sitting around the fire playing some kind of game. They looked like outlaws, certainly not to be trusted. He had no other choice but to approach them. He was hungry, tired, had not slept or had water for few days...

<div id="picture"><img width="400" src="files/images/outlaws.jpg">
	</div>

Mr Anderson asked for water but the characters of the camp told him that in the desert nobody gives you water; you have to earn it. Next they invited him to join them in a game of dudo, a game similar to poker but played with dice. Mr Anderson knew nothing about the game but quickly realized after a bit of observation that it is all based on probability theory. After few rounds, he realized that all the players were using strategies derived from calculating probabilities based on commutative probability and extreme value distributions. Thanks to a little project he did with Prof. Provlief he had a small computer program that used Markov Chain Monte Carlo to calculate complicated integrals of probability distributions. Luckily the program that he used for the project was still on his Texas Instrument calculator that had survived the crash. Quickly he modified the code and soon after he started to win round after round. As winner of the game, he left the camp with water, food and a horse. He was feeling good about himself and all the time he spent with Prof. Provleif. Again it had saved his life.

On the horse now, he continued his journey trying to get out of the desert and still determined to make a lot of money before returning home. After a day of riding he came upon a little mining village. He was ecstatic to eventually be in a place where he could eat a hot meal, take a bath and rest a little before continuing on his quest to become rich. He checked into a local hotel and went for dinner at a local bar called the Empanadillas Caliente. He sat down at the bar next to a big strong man -- Mr. Jimenez was his name. He asked Mr. Jimenez what he was doing and was told that he was a mining engineer looking for iodine. Apparently there was a lot of demand for this mineral as a result of the rapidly expanding business of photography (iodine is an essential ingredient in film). Mr Jimenez was looking to find ways to extract iodine from nitrate but so far had only found failure. Nitrate and iodine were naturally to be found in mixed pools but nobody knew how to locate the pools. People could only use sonars to find the mineral cores but sonar wasn't sufficient to distinguish between the two. After couple of piscos, he again went back to his memory banks of the lectures of Prof. Provlief to recall that expectation maximization could be used to deduce the values of things not directly observed. He continued conversing with Mr. Jimenez and quickly realized that the methods he learned can be applied to the problem of extracting iodine. This led to a vivid discussion between the two technically savvy people while they downed a lot of piscos. By the morning they had figured out the likelihood function, and on his little calculator he was able to run the code and estimate the amount of iodine vs nitrate on many of the possible reservoirs.

In the next few months they worked intently to perfect the method and started a new iodine exploration company. Their new company started making money and soon Mr. Anderson had enough money to feel very good about things. His dream was coming together - finally.

Sadly it turns out that he had started to feel comfortable too soon. His relationship with Mr. Jimenez went bad after the first few months. It turned out that Mr. Jimenez was not so honest with his dealings and he made Mr. Anderson sign a bad contract. Mr. Anderson, trusting that Mr. Jimenez was his friend and had his best interests in mind, did this without the proper due diligence. Mr. Anderson ended up up with only a dozen thousand dollars and completely out of the company. Totally disappointed but at least with a little cash in his pocket he took the first plane out of South America to London.

His first priority was to hit the local pub. There he met Mr. Muray who was working at a financial firm in equity trading. Mr. Anderson and Mr. Muray become friends over a warm beer at the pub. Mr Muray was talking away about the intricacies of trading and again Mr. Anderson once again connected back to the ever-influential Prof. Provlief's lectures on Hidden Markov Models and how to simulate sequential events of hidden variables. Quickly he pulled his little computer and started calculating. Long story short Mr. Anderson now lives a comfortable existence as a multimillioner trading on wall street.
