# Perfect Sampling

In this lecture, we will discuss "Perfect Sampling" algorithms. These are relatively recent developments, but it has received an enormous amount of attention when the "coupling from the past" algorithm was suggested by Propp and Wilson in 1996, and it still remains as an active research area. The subject has fascinated many researchers, partly because the algorithm itself is really cool, but also because it has great potential to do wonderful things in MCMC and other contexts. 

Before we start, let's review the big picture first. We have already learned in the previous lectures that Markov chain Monte Carlo algorithms sample from a given target distribution by constructing a Markov chain whose equilibrium distribution is the target distribution. We have seen a number of ways to construct such Markov chains&mdash;Metropolis-Hastings, Gibbs, Slice, Hamiltonina MCMC, etc. Once we have such a Markov chain, we can study the target distribution using the samples obtained by running the Markov chain for a long time, because if we run a Markov chain for a long enough time, the finite time distribution will be sufficiently close to the equilibrium distribution. The question is, of course, how long is long enough? To (incompletely) answer this question, we also have seen a few diagnostics. However, there's no general method to give a definite answer to this question, and determing the convergence of Markov chains remains to be a notoriously hard problem. 

The class of algorithms we are going to discuss today approaches this problem from a different perspective. Instead of approximating the equilibrium distribution of a given Markov chain and reducing or bounding the approximation error, these algorithms try to sample from the exact equilibrium distribution of the Markov chain with perfect precision. Due to this grand ambition, people like to call them "perfect simulation" or "perfect sampling" algorithms. Less exciting but more informative terms "exact simulation" or "exact sampling" are also frequently used.

We will focus on the classic perfect sampling algorithm&mdash;Propp and Wilson's coupling from the past (CFTP).


## The Goal

Given a Markov chain $\{x^{(i)}\}$ such that $x^{(i)} \Rightarrow x^{(\infty)}$, sample from the equilibrium distribution $\pi$ of $x^{(\infty)}$ 

## The Classic: Coupling from the Past (CFTP)

As any english speaker can easily guess from the name, coupling from the past algorithm generates exact samples from the equilibrium distribution of a Markov chain by (virtually) running the Markov chain from its infinite past. But, what's mysterious in the name is the word "coupling."

### What is Coupling?

In probability, coupling refers to the construction of a probability space where two random objects are defined simultaneously. In perfect simulation context, the coupling referes to the coalescence of paths from all possible starting points. Here is an example that we will follow throughout this lecture.

### An Example: Random walk with boundary

Let's consider the following Markov chain evolving in state space $S = \{0, 1, \cdots, M\}$:

\begin{align}
x^{(n+1)} &= \min\big( \max \big(x^{(n)} + \xi^{(n+1)}, 0\big), M\big)
\end{align}

where $\xi^{(n)}$ is iid with $P(\xi^{(n)} = -1) = P(\xi^{(n)} = 1) = 1/2$. This is a simple random walk that evolves within the boundaries at $0$ and $M$.

We can construct the coupling (or force the coalescence) by using common random noise $\xi^{(n)}$ for all paths.



### Coupling (Coalescence) of paths

Since we will consider multiple versions (i.e., realized paths) of the Markov chain each of which starts from different points in the state space, it will be convenient to introduce a new notation: let $x^{(n)}(s)$ denote the Markov chain that has started from state $s \in S$ (at time $0$). Once two or more versions of these paths meet each other, we can see from the above dynamics (recursion) of the Markov chain that those paths will stay together forever afterwards. That is, if $x^{(k)}(s_1) = x^{(k)}(s_2)$ for some $k$, $x^{(k+i)}(s_1) = x^{(k+i)}(s_2)$ for all $i\geq 0$. The python code below illustrates such coupling/coalescence. 

```python
def generate_noise(l):
    return 2*bernoulli.rvs(0.5, size=l)-np.ones(l)
```

```python
def run(noises,low_bnd,up_bnd,start_pt):
    x = np.zeros(len(noises)+1)
    x[0] = start_pt
    for i,noise in enumerate(noises):
            x[i+1] = min( max( x[i]+noise, low_bnd ), up_bnd )
    return x
```

```python
l=50
M=10
noises = generate_noise(l)
print noises
fig = plt.figure()
for i in range(M+1):
    path = run(noises,0,M,i)
    plt.plot(path)
```

### Analytic Solution

We have intentionally chosen a tractable example to have a reference distribution. The transition matrix of the Markov chain, for example in case $M = 3$, i.e., $\{x^{(n)}\}$ has state space $S = \{0,1,2,3\}$, is 

$$ P = 
\left(\begin{array}{cccc}
1/2 & 1/2 & 0   & 0\\ 
1/2 & 0   & 1/2 & 0\\
0   & 1/2 & 0 & 1/2\\
0   & 0   & 1/2 & 1/2
\end{array}\right).$$

From the form of the transition matrix $P$, we can check that $\pi = (1/4, 1/4, 1/4, 1/4)$ satisfies the steady-state equation

$$\pi P = \pi.$$

Hence, the equilibrium distribution of $\{x^{(n)}\}$ is uniform distribution on state space. This is true for any $M$. This can be also seen by computing the power of the transition matrix $P$.

```python
P = np.matrix([[0.5, 0.5, 0,   0  ],\
               [0.5, 0,   0.5, 0  ],\
               [0,   0.5, 0,   0.5],\
               [0,   0,   0.5, 0.5]])
print P; print
print P**2;print
print P**5;print
print P**10;print
print P**20;print
print P**50;print
print P**200;print
```

### Coupling from the Past

Yet another notation: we will denote the Markov chain that has started from time $m$ at state $s$ with $x^{(n)}(s;m)$.

The classic coupling from the past algorithm proceeds as follows.

1. Set starting time $n = -1$.
2. Initialize our Markov chain $\{x^{(i)}\}$ at time $n$ from each point $s$ of the state space $S$.
3. Generate a random noise $\xi^{(n+1)}$.
4. Using the (common) random noise $\xi^{(n+1)}, \xi^{(n+2)}, \cdots, \xi^{(0)}$ generated in step 3, run each Markov chain upto time 0.
5. If all the paths have coupled (coalesce) by time 0, return the common value $x^{(0)}$.
    If not, set $n = n-1$ and go to step 2.



### Implementation and Numerics

Let's see if CFTP works for our bounded random walk example. Note that in the following implementation, we didn't have to simulate the process from all possible starting points because of the monotonicity of our random walk example. We only had to monitor two versions: one started from $M$ and the other one started from $0$. If those two versions couple, then we know that all the other versions were sandwiched between them, and hence, every path has coalesced. Also, it should be noted that it is not necessary to detect the coalescence from the most recent starting points. In other words, we don't have to scan every starting point -1, -2, -3,$\ldots$. Instead, we can scan backward using binary search, i.e., doubling the number of time steps we go back (-1, -2, -4, -8,$\ldots$) instead of increasing by 1. Finally, we also note that we don't have to start from $t = -1$, because there's no way the backward chain coalesce in one step; it requires at least $M$ steps to reach the bottom from the top. Therefore, we started at $t=-M$ instead of $t=-1$.

```python
def cycle(innovs,low_bnd,up_bnd):
    upper = np.zeros(len(innovs)+1)
    lower = np.zeros(len(innovs)+1)
    upper[0] = up_bnd
    lower[0] = low_bnd
    for i,innov in enumerate(innovs):
            upper[i+1] = min(max(upper[i] + innov,low_bnd),up_bnd)
            lower[i+1] = min(max(lower[i] + innov,low_bnd),up_bnd)
    return upper,lower
```

```python
def perfect_sampling(M):
    l = M
    noises = generate_noise(l)
    while True:
        low = run(noises,0,M,0)
        up  = run(noises,0,M,M)
        if low[l] == up[l]: return low[l];
        noises = np.concatenate((generate_noise(l),noises),axis=0)
        l = 2*l
```

```python
N = 10000
samples = np.zeros(N)
for i in range(N):
    samples[i] = perfect_sampling(10)
plt.hist(samples,bins=11)
```

Fantastic! CFTP indeed works! It should be noted that, here it was important to 

* use common random numbers for all paths
* extend the common path backward (not forward) until coupling takes place
* re-use the random numbers to run the backward processes
* sample at time 0

At first glance, however, it may not be clear why these are important and why CFTP works. Let's see what can go wrong. 

### What if we do not re-use random numbers? 

The code below illustrates what happens if we generate new random noise rather than reusing the old random noise when we go back to the further past. The only differences between "perfect_sampling" above and "without_reuse_sampling" below are the last two lines.

```python
def without_reuse_sampling(M):
    l = M;
    noises = generate_noise(l)
    while 'true':
        up,low = cycle(noises,0,M)
        if low[l] == up[l]:
            return low[l];
        l = 2*l
        noises = generate_noise(l)

        
N = 10000
samples = np.zeros(N)
for i in range(N):
    samples[i] = without_reuse_sampling(10)
plt.hist(samples,bins=11)
```

This looks quite different from the uniform distribution. 

### What if we run the process forward?

One may wonder "what if we do the same thing, but search for the coalescence in forward direction?" 

```python
def forward_sampling(M):
    l = M;
    innovs = generate_noise(l)
    while 'true':
        up,low = cycle(innovs,0,M)
        if low[l] == up[l]:
            return low[l];
        l = 2*l
        innovs = innovs = np.concatenate((innovs,generate_noise(l)),axis=0)

        
N = 10000
samples = np.zeros(N)
for i in range(N):
    samples[i] = forward_sampling(10)
plt.hist(samples,bins=11)
```

The answer is no.

### Why does CFTP work?

In general, a Markov chain can be written as a recursion. For example, in case of our random walk example, the recursion looks as follows:

$$ x^{(n+1)} = r(x^{(n)}, \xi^{(n+1)})$$

where $r(s,t) = \min\big( \max\big( s+t, 0\big), M\big)$.

If we denote $r(\cdot, \xi^{(n)})$ with $\varphi_i(\cdot)$, then each $\varphi_i$ is a random function, and $x^{(n)}$ can be written as an iterated random functions. 

\begin{align}
x^{(1)} &= r(x^{(0)}, \xi^{(1)}) = \varphi_1(x^{(0)})\\
x^{(2)} &= r(x^{(1)}, \xi^{(2)}) = \varphi_2(x^{(1)}) = \varphi_2\circ \varphi_1(x^{(0)})\\
&\vdots\\
x^{(n)} &= \varphi_n\circ \cdots \circ \varphi_1(x^{(0)})
\end{align}

Note that since $\xi^{(i)}$-s are iid, $\varphi_i(\cdot) = r(\cdot, \xi^{(i)})$ are iid, and hence,

\begin{align*}
x^{(n)} 
&= \varphi_n\circ \cdots \circ \varphi_1(x^{(0)})\\
&\stackrel{\mathcal D}{=} \varphi_1\circ \cdots \circ\varphi_n(x^{(0)})\\
&\stackrel{\mathcal D}{=} \varphi_{-1}\circ \cdots \circ \varphi_{-n}(x^{(0)})\\
&\triangleq \tilde x^{(n)}
\end{align*}

where $\tilde x^{(n)}$ is the "backward process." This implies that

\begin{align*}
x^{(\infty)} \stackrel{\mathcal D}{=} \tilde x^{(\infty)}
\end{align*}

and hence if we can generate exact samples of $\tilde x^{(\infty)}$, then we are generating samples from $\{x^{(n)}\}$'s exact equilibrium distribution. 

At this point, it might not be clear yet why sampling from $\tilde x^{(\infty)}$ is easier than sampling from $x^{(\infty)}$. It is worthwhile noticing a couple of important differences between the forward process and the backward process:

* $\tilde x^{(n)}$ is not a Markov chain. We can not compute $\tilde x^{(n)}$ from $\tilde x^{(n-1)}$ and $\varphi_{-n}$ alone. To compute $\tilde x^{(n)}$ we need all the previous $\varphi_{i}$-s (and the starting point, of cousre).
* If $\varphi_{-1}\circ\cdots\circ\varphi_{-k}$ is constant, $\tilde x^{(k)} = \tilde x^{(k+1)} = \cdots = \tilde x^{(\infty)}$ for all $i\geq 0$.

The second property makes it possible to generate $\tilde x^{(\infty)}$ in finite time (that is, $k$). In other words, if we let 

$$T \triangleq \sup\{s\leq 0: \varphi_{-1}\circ\cdots\circ\varphi_{-k}(\cdot) \text{ is constant}\},$$ 

then $\tilde x^{(T)} = \tilde x^{(\infty)}$ w.p. 1.

Therefore, if 

* we can construct a probability space where $\varphi_{-1}\circ\varphi_{-2}\circ\cdots\circ\varphi_{-n}$ becomes constant (coalescence) with probability 1;
* we can detect such coalescence,

then, CFTP algorithm should work.


### Challenges for Routine Applications

Although it was deceptively easy in case of our random walk example, constructing a coupling so that all paths coalesce at reasonable time is not a trivial task. Even for a very simple Markov chain with two states, we can construct a coupling (a joint probability space) where two paths from the two starting points never coalesce. Suppose that the state space $S = \{0,1\}$, and the perturbations $\xi_i$ are iid Bernoulli with success probability 1/2. Let 

$$\varphi_i(s) = r(s,\xi_i) = \delta_{s\xi_i}.$$

where $\delta_{st}$ is the Kronecker delta function. Then, 

$$\varphi_i(s) \neq \varphi_i(t) \quad \text{iff} \quad s \neq t$$ 

and hence 

$$\varphi_1\circ \cdots \circ \varphi_n(0) \neq \varphi_1\circ \cdots \circ \varphi_{n}(1)$$

w.p. 1 for all $n$. That is, for $x^{(n)}(s) = \varphi_n\circ \cdots \circ \varphi_1(s)$ which has stationary distribution $\pi(0) = \pi(1) = 1/2$, the backward chain does not converge in pathwise sense.


Even if we are successful in constructing a reasonable coupling, detecting the coalescence can be challenging. For example, if the state space is extremely large (e.g., Ising model) which is, the norm rather than the exception in perfect simulation context&mdash;for small state spaces we can easily solve for the steady state equation $\pi P = \pi$ to directly compute the equilibrium distribution. The problem becomes even harder if we consider (coutably) infinite state spaces, or even continuous state spaces. As we have seen in the previous implementation of our random walk example, the monotonicity can help a lot, and there are other clever ways developed to detect the coalescence even without monotonicity. In fact, the class of processes for which we can construct CFTP algorithm has been identified.

###  Necessary and Sufficient Condition for Coupling from the Past

Foss and Tweedie proved that successful CFTP algorithms can be constructed if and only if the chain is uniformly ergodic. The sufficient condition can be proved by following the construction of multigamma coupling.


### Multi-gamma Sampler

If a Markov chain on a state space $S$ with transition kernel $p(x,dy)$ is uniformly ergodic, then there exists a measure $\nu$ and constants $m\geq 1$ and $0< \beta\leq 1$ such that

$$P^m(s,\cdot) \geq \beta \nu(\cdot), \qquad \forall s \in S$$

This effectively means that we can construct a probability space where $\varphi_i\circ\varphi_{i+1}\circ\cdots\circ\varphi_{i+m-1}(\cdot)$ is constant with positive probability for each $i$. In view of what we have discussed in section "Why does CFTP work?," we can see why uniform ergodicity is a sufficient condition for successful CFTP algorithm.
